```python
# In this notebook, I am loading a model that I trained on a fashion mnist dataset
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset and pick an image
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

```

```python
image = x_test[0]
plt.imshow(image, cmap='gray')
plt.show()
```

```python
# load the model from the models folder
model = tf.keras.models.load_model('./models/CNN1.h5')
```

```python
data_classes = np.array(['T-Shirt', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot'])
img = np.array([image])
single_prediction = model.predict(img)
print(f'Prediction: {data_classes[np.argmax(single_prediction)]}')
print(f'Confidence: {((np.max(single_prediction)*100)):.2f}%')
```
